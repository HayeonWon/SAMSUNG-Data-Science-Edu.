### [예제1] [] 연산 indexing:  Series

**Series Indexing**
* sr[ ] : 하나의 value -> scalar 반환
* sr[[...]] : 여러 value -> series 반환
* sr[:] : slicing -> series 반환
* sr[::] : slicing(step) -> series 반환

```python
# [1]

sr1 = pd.Series([10,20,30,40],index=['A','B','C','D'])

r1 = sr1['B']
r2 = sr1[1]
r3 = sr1[['A','C']]  # ['A','C'] 리스트로 indexing
r4 = sr1[[0,2]]
r5 = sr1['A':'C']  # slicing
r6 = sr1[0:3]
sr1; r1; r2; r3; r4; r5; r6
```

```python
# [2]

sr2 = pd.Series([10,20,30,40],index=[1,2,3,4])

r7 = sr2[1]
r8 = sr2[1:3]
r9 = sr2[::2]
sr2; r7; r8; r9
```

### [예제2] [ ] 연산 indexing: DataFrame

**DataFrame Indexing**
* df[ ] : 하나의 column -> series 반환
* df[[...]] : 여러 column -> DF 반환
* df[:] : 행 slicing -> DF 반환
* df[::] : 행 slicing(step) -> DF 반환

```python
# [1]
# 컬럼 선택
# label만 사용 (위치번호 X)
# [column label]로 indexing

df = pd.DataFrame(np.arange(12).reshape(4,3), index=list('ijkl'), columns=list('ABC'))
r1 = df['A']
r2 = df[['A']]
r3 = df[['A','C']]
disp('df'); r1; disp('r2', 'r3')
```

```python
# [2]
# 행 선택
# 위치번호와 label 사용
# 반드시 [:]로 slicing (위치번호 OR label 사용)

df = pd.DataFrame(np.arange(12).reshape(4,3), index=list('ijkl'), columns=list('ABC'))

r4 = df['i':'k']
r5 = df[0:3]
r6 = df[::2]
r7 = df['k':'i':-1]
disp('df', 'r4', 'r5', 'r6', 'r7')
```

### [예제4] .loc과 .iloc을 이용한 indexing

* .loc은 "label" 기반 indexing
* .iloc은 "위치번호" 기반 indexing

* df.loc[idx] / df.iloc[idx] : idx에 따라 하나 또는 여럿 선택
* df.loc[row_idx] / df.iloc[row_idx] : row만 선택, 모든 col
* df.loc[row_idx, col_idx] / df.iloc[row_idx, col_idx] : row와 col 둘 다 선택

```python
# [.loc => r1, r2, r3, r4]

df1 = pd.DataFrame(np.arange(9).reshape(3,3), index=list('ijk'), columns=list('ABC'))

r1 = df1.loc['i', 'B'] # i행 B열
r2 = df1.loc['i', :]   # i행 모든열
r3 = df1.loc['i']      # i행 모든열
r4 = df1.loc[:, 'B']   # 모든행 B열
df1; r1; r2; r3; r4
```

```python
# [.loc => r5, r6, r7,r8,r9]

df1 = pd.DataFrame(np.arange(9).reshape(3,3), index=list('ijk'), columns=list('ABC'))

r5 = df1.loc[['i','k']]    # i행, k행
r6 = df1.loc['i':'k']      # i~k행
r7 = df1.loc[:, ['A','C']] # A열, C열
r8 = df1.loc[:, 'A':'C']   # A~C열
r9 = df1.loc[['i','k'], 'A':'C':2] # i,k행 A,C열
disp('df1', 'r5','r6', 'r7','r8','r9')
```

```python
# [.loc ==> r10]

df2 = pd.DataFrame(np.arange(9).reshape(3,3), index=[1,2,3], columns=[5,6,7])

r10 = df2.loc[1:2] # 1,2 행
disp('df2', 'r10')
```

```python
# [.iloc => r1, r2, r3, r4]

df1 = pd.DataFrame(np.arange(9).reshape(3,3), index=list('ijk'), columns=list('ABC'))

r1 = df1.iloc[0, 1] # 1번째 행, 모든 열
r2 = df1.iloc[0, :] # 1번째 행, 모든 열
r3 = df1.iloc[0]    # 1번째 행, 모든 열
r4 = df1.iloc[:, 1] # 모든 행, 2번째 열
df1; r1; r2; r3; r4
```

```python
# [.iloc => r5, r6, r7,r8,r9]

df1 = pd.DataFrame(np.arange(9).reshape(3,3), index=list('ijk'), columns=list('ABC'))

r5 = df1.iloc[[0,2]]     # 1,3번 행  -> 슬라이싱이 아니므로 리스트로 인덱싱 ★★★★★
r6 = df1.iloc[0:3]       # 1~3번째 행 
r7 = df1.iloc[:, [0,2]]  # 모든 행, 1,3번 열 ★★★★★
r8 = df1.iloc[:, 0:3]    # 모든 행, 1~3번 열 ★★★★★
r9 = df1.iloc[[0,2], 0:3:2]  # 1,3번 행 / 1,3번 열
disp('df1', 'r5','r6', 'r7','r8','r9')
```

```python
# [.iloc ==> r10]

df2 = pd.DataFrame(np.arange(9).reshape(3,3), index=[1,2,3], columns=[5,6,7])

r11 = df2.iloc[1:2] # 2번 행
disp('df2', 'r11')
```
