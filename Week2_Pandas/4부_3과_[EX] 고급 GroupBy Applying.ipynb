### [예제1] gb.apply() 메서드의 이해

```python
# [1-1]

d = {'class':list('ABAB'),'kor':[60,80,100,90],'eng':[80,50,60,70]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])

def f1(x):
    print(x)
    return x.sum()

r1 = df.groupby('class').apply(f1)
disp('df', 'r1')
```

```python
# [1-2]

d = {'class':list('ABAB'),'kor':[60,80,100,90],'eng':[80,50,60,70]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])

def f1(x):
    print(x)
    return x.sum()

r2 = df.groupby('class')['kor'].apply(f1)
df; r2
```

```python
# [2-1]

d = {'class':list('ABAB'),'kor':[60,80,100,90],'eng':[80,50,60,70]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])

def f2(x, n, col):
    return x.nlargest(n, col)

r3 = df.groupby('class').apply(f2, n=1, col='kor')  # class별 kor가 가장 큰 행만 추출
disp('df', 'r3')
```

```python
# [2-2]

d = {'class':list('ABAB'),'kor':[60,80,100,90],'eng':[80,50,60,70]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])

def f3(x):
    return x.mean(axis=1,numeric_only=True).describe()

r4 = df.groupby('class').apply(f3)
disp('df', 'r4')
```

### [예제2] gr.transform 메서드의 이해

* transform은 원본형태로 그대로 반환해줌

```python
# [1-1]

d = {'class':list('ABAB'),'kor':[40,100,80,90],'eng':[85,75,65,95]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])

def f1(x):
    print(x)
    return x.max()

r1 = df.groupby('class').transform(f1)
disp('df', 'r1')
```

```python
# [1-2]

d = {'class':list('ABAB'),'kor':[40,100,80,90],'eng':[85,75,65,95]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])

def f1(x):
    print(x)
    return x.max()

r3 = df.groupby('class')['kor'].transform(f1)
df; r3
```

```python
# [2]

d = {'class':list('ABAB'),'kor':[40,100,80,90],'eng':[85,75,65,95]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])
df['avg'] = df.mean(axis=1, numeric_only=True)
df

df['class_avg'] = df.groupby('class')['avg'].transform('mean')
df['class_dev'] = df['avg'] - df['class_avg']
df
```

```python
d = {'class':list('ABAB'),'kor':[40,100,80,90],'eng':[85,75,65,95]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])
df['avg'] = df.mean(axis=1, numeric_only=True)
df

df.groupby('class')['avg'].mean()               # class별 avg컬럼 평균
df.groupby('class')['avg'].transform('mean') # class별 avg컬럼 평균 원래 df형태로 반환
```

### [예제3] gr.agg() 메서드의 이해

```python
# [1-1]

d = {'class':list('ABAB'),'kor':[40,100,80,90],'eng':[85,75,65,95]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])

def f1(x):
    print(x)
    return x.max()

r1 = df.groupby('class').agg(f1)
disp('df', 'r1')
```

```python
# [1-2]

d = {'class':list('ABAB'),'kor':[40,100,80,90],'eng':[85,75,65,95]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])

def f1(x):
    print(x)
    return x.max()

r2 = df.groupby('class')['kor'].agg(f1)
df; r2
```

```python
# [2-1]

d = {'class':list('ABAB'),'kor':[40,100,80,90],'eng':[85,75,65,95]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])

gr = df.groupby('class')
r3 = gr['kor'].agg(['max', 'min'])
r4 = gr['kor'].agg([('kormax','max'), ('kormin','min')])
disp('df', 'r3', 'r4')
```

```python
# [2-2]

d = {'class':list('ABAB'),'kor':[40,100,80,90],'eng':[85,75,65,95]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])

gr = df.groupby('class')
r5 = gr.agg({'kor':'max','eng':'min'})
r6 = gr.agg(kormax=('kor','max'), engmin=('eng','min'))
disp('df', 'r5', 'r6')
```

### [예제4] gr.filter() 메서드의 이해

```python
d = {'class':list('ABAB'),'kor':[40,100,80,90],'eng':[85,75,65,95]}
df = pd.DataFrame(d, index=['kim','lee','park','song'])
df['avg'] = df.mean(axis=1)

def f1(x):
    print(x)
    return x['avg'].mean() >= 80

def f2(x):
    return x.mean() >=80

r1 = df.groupby('class').filter(f1)
# r2 = df.groupby('class').filter(f2) # error

disp('df', 'r1')
```

### [예제5] 그룹 집계 활용 : titanic

```python
# [1-1]

df = pd.read_csv('data/titanic.csv', usecols=['pclass','survived','sex','age']) 
print(df.groupby(['pclass','sex'])['age'].mean())
```

```python
# [1-2]

df = pd.read_csv('data/titanic.csv', usecols=['pclass','survived','sex','age']) 

gb = df.groupby(['pclass','sex'], group_keys=False)
df['age2'] = gb['age'].apply(lambda x: x.fillna(x.mean()))
df.head(20)
```

```python
# [2] 

df = pd.read_csv('data/titanic.csv',usecols=['pclass','survived','sex','age']) 

gb = df.groupby(['pclass', 'sex'])
r2 = gb['survived'].agg([('all','count'),('alive','sum')])  
r2
```
